# New Geometry 1

```python
(ggplot(data=df1, mapping=aes(x='income', y='population', color ='factor(life_exp)'))
+ geom_jitter(alpha=0.3)
+ geom_violin(alpha=0, color="0.7")
+ scale_y_log10()
+ facet_wrap('four_regions')
+ theme(figure_size=(10, 6),
    aspect_ratio=1/2    )
)
```

This geometry is not the best representation of this data. Although it includes all four variables, it doesn't convey the same story as clearly as the bubble graph. I think the life_exp bubbles and it's relationship with the other variables can be a little confusing as well.
