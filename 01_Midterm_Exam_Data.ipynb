### _geoloc

```python
type(park['_geoloc'])
```

```python
pak = list(park['_geoloc'])

one_p = list(map(lambda x: x.replace("{",""), pak))
sec_p = list(map(lambda x: x.replace("}",""), one_p))
no_apost = list(map(lambda x: x.replace((("'")),""), sec_p))
string = str(no_apost)
sec_apost = string.replace("'","")
no_colon = sec_apost.replace(":","")
no_colon
#re.findall(r"(lat\d", sec_apost)
```

```python
latitudes = [float(match.group(1)) for match in re.finditer(r'lat ([-+]?\d*\.\d+)', no_colon)]
longitudes = [float(match.group(1)) for match in re.finditer(r'lng ([-+]?\d*\.\d+)', no_colon)]

# Create a DataFrame
data = pd.DataFrame({
    'Latitude': latitudes,
    'Longitude': longitudes
})
data
```

```python
park = park.join(data)
```

## It’s reasonable to think we might also want to search for hikes according to climate. Therefore, we will scrape the average temperatures for each U.S. state from the following website:

https://www.extremeweatherwatch.com/us-state-averages
Write a function that takes one optional arguent, month.



```python
# scraping
import requests
response = requests.get("https://www.extremeweatherwatch.com/us-state-averages")


from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
```

```python
table = soup.find("table", class_ = 'bordered-table sort-table')
table
```

```python
rows = []

# iterate over all rows in the faculty table
for states in table.find_all("tr")[1:]: # tr is table rows

    # Get all the cells (<td>) in the row.
    cells = states.find_all("td") # td is cells in rows

    # The information we need is the text between tags.

    # Find the the name of the faculty in cell[0]
    # which for most faculty is contained in the <strong> tag
    state_name = cells[0].find("a") or cells[0]
    State = state_name.text

    # Find the office of the faculty in cell[1]
    # which for most faculty is contained in the <a> tag
    high_temp = cells[1].find("td") or cells[1]
    High = high_temp.text

    # Find the email of the faculty in cell[3]
    # which for most faculty is contained in the <a> tag
    low_temp = cells[3].find("td") or cells[3]
    Low = low_temp.text


    # Append this data.
    rows.append({
        "State": State,
        "High Average Temperature": High,
        "Low Average Temperature": Low
    })
rows
```
