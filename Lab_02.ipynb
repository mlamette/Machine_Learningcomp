# In which California regions is the price of organic versus conventional avocados most different? Support your answer with a few summary statistics AND a visualization?

Answer: San Francisco

```python
differences = pd.crosstab(cali_regions['geography'], columns = cali_regions['type'], values = cali_regions['average_price'], aggfunc = 'mean')
differences['price_difference'] = differences['conventional']-differences['organic']
differences['price_difference'] = differences['price_difference'].abs() #make the differences positive
differences = differences.reset_index()
differences.columns.name = None
differences
```

```python
#visualization
#Los Angeles
la = cali_regions[cali_regions['geography']=='Los Angeles']
la = la.loc[:,['average_price','type','geography','year']]
o = la[la['type']=='organic']
c = la[la['type']=='conventional']
ladiff = o.merge(c, on=['year'])
ladiff['differences'] = ladiff['average_price_x']-ladiff['average_price_y']
ladiff

#Sacramento
sac = cali_regions[cali_regions['geography']=='Sacramento']
sac = sac.loc[:,['average_price','type','geography','year']]
o = sac[sac['type']=='organic']
c = sac[sac['type']=='conventional']
sacdiff = o.merge(c, on=['year'])
sacdiff['differences'] = sacdiff['average_price_x']-sacdiff['average_price_y']
sacdiff

#San Diego
sd = cali_regions[cali_regions['geography']=='San Diego']
sd = sd.loc[:,['average_price','type','geography','year']]
o = sd[sd['type']=='organic']
c = sd[sd['type']=='conventional']
sddiff = o.merge(c, on=['year'])
sddiff['differences'] = sddiff['average_price_x']-sddiff['average_price_y']
sddiff

#San Francisco
sf = cali_regions[cali_regions['geography']=='San Francisco']
sf = sf.loc[:,['average_price','type','geography','year']]
o = sf[sf['type']=='organic']
c = sf[sf['type']=='conventional']
sfdiff = o.merge(c, on=['year'])
sfdiff['differences'] = sfdiff['average_price_x']-sfdiff['average_price_y']
sfdiff

differences = pd.concat([ladiff,sacdiff,sddiff,sfdiff],ignore_index=True)
differences.rename(columns = {'differences':'price_diff_ovsc'}, inplace=True)
differences
```

```python
shift = 0.1
mean_values = differences

def alt_sign(x): #created a function to move it
    "Alternate +1/-1 if x is even/odd"
    return (-1) ** x

m1 = aes(x=stage('geography_x', after_scale='x+shift*alt_sign(x)'))

d = (ggplot(differences, aes(x = 'geography_x',y ='price_diff_ovsc', fill = 'geography_x'))
+ geom_violin(m1, style = 'left-right')
+ geom_boxplot(width = 0.1, alpha = 0.7, size = 0.65)
+ xlab("California Regions")
+ ylab("Price Difference")
+ labs(fill = "Regions")
+ ggtitle("Average Price Differences Between Conventional and Organic Avocados")
+ scale_y_continuous(breaks=np.arange(0, 2.6, 0.5), limits=[0, 2.5])
+ scale_fill_manual(values=['dodgerblue', 'darkorange', 'mediumseagreen','coral'])
+theme(
        axis_line=element_line(size=1, colour="black"),
        panel_grid_major=element_line(colour="#d3d3d3"),
        panel_grid_minor=element_blank(),
        panel_border=element_blank(),
        panel_background=element_blank(),
        plot_title=element_text(size=15, family="monospace",
                                face="bold"),
        text=element_text(family="monospace", size=11),
        axis_text_x=element_text(colour="black", size=10),
        axis_text_y=element_text(colour="black", size=10),
    )
+ theme(figure_size = (12,6)))
d

#still need the multiple observations from the dataset
```

```python
import pandas as pd
import random

#Chatgpt assisted - I prompted it to make me a dataset

# Define the range of years, months, and geographies
years = list(range(2015, 2021))
months = list(range(1, 13))
geographies = ['Los Angeles', 'Sacramento', 'San Diego', 'San Francisco']

# Create an empty DataFrame to store the data
data = []

# Generate random median house prices for each combination
for year in years:
    for month in months:
        for geography in geographies:
            median_price = round(random.uniform(300000, 800000), 2)  # Random median price between $300,000 and $800,000
            data.append([year, month, geography, median_price])

# Create a DataFrame from the generated data
df = pd.DataFrame(data, columns=['Year', 'Month', 'Geography', 'Median_House_Price'])

# Display the first few rows of the dataset
print(df.head())
```

```python
b = pd.melt(cali_regions, id_vars = ['geography','type'], value_vars=['small/medium','large','extra large'])
b = b.rename(columns={'variable':"plu",'value':'total avos sold'})
ft = b.groupby(['geography','type','plu']).agg({'total avos sold':'sum'}).reset_index()
ft = ft.rename(columns = {'geography':'Geography'})
ft
avohouses = ft.merge(df,on = 'Geography')
avohouses
```

```python
p = (ggplot(avohouses, aes(x = 'Median_House_Price', y= 'total avos sold', color = 'Geography'))
+ geom_line(aes(size='plu'))
+ facet_wrap(['type'])
+ labs(
    x = 'Median House Price',
    y = 'Total Avocados Sold',
    title = 'House Prices v. Avocado sales'
)
+ scale_x_continuous(
        breaks=[300000, 400000, 500000, 600000, 700000, 800000],
        labels = ["300k","400k","500k","600k","700k","800k"]
    )
+ theme(
        axis_line=element_line(size=2, colour="lightgray"),
        panel_grid_major=element_line(colour="#d3d3d3"),
        panel_grid_minor=element_blank(),
        panel_border=element_blank(),
        panel_background=element_blank(),
        plot_title=element_text(size=15,
                                face="bold"),
        text=element_text(size=11),
        axis_text_x=element_text(colour="grey", size=10, angle=45, hjust = 1),
        axis_text_y=element_text(colour="grey", size=10),
        strip_background=element_rect(fill='darkgrey'),
        strip_text=element_text(colour = 'white'))
+ theme(figure_size = (12,6)))
p
```
