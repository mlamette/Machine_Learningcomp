# Packages

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer,make_column_selector
from sklearn.preprocessing import StandardScaler, OneHotEncoder, FunctionTransformer
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.model_selection import cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.tree import export_graphviz
from IPython.display import display, Image
import graphviz
from sklearn import tree
from sklearn.metrics import f1_score, confusion_matrix
from plotnine import *
from sklearn.svm import SVC
```

# Data

```python
ct = pd.read_csv("https://www.dropbox.com/s/s2a1uoiegitupjc/cannabis_full.csv?raw=1")
ct = ct.dropna()
ct.head()
```

```python
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

# Part One: Binary Classification

I am choosing ROC AUC as my metric because it is the best metric that gives you the overall picture, the overall evaluation of the models. If there was a specific question or specific business question we are trying to answer, I would consider looking at other metrics.

For example, if I was concerned with specifically the true postive rate I would focus more on recall or sensitivity.

```python
ctnew = ct[(ct['Type']=='sativa') | (ct['Type']=='indica')]
```

```python
X = ctnew.drop(['Type','Strain', 'Effects','Flavor'], axis = 1)

y = ctnew['Type']
```

```python
# Confusion Matrix function

def confusion_matrix_pic(model_name,pipeline, X, y):
    '''
    Makes a confusion matrix visualization

    Parameter
    --------
    model_name : str
    fitted_model: fitted pipeline
    y : panda series of target variable
    y_pred: panda series from fitted_model.predict(X)

    Returns
    -------
    ggplot confusion matrix using plotnine

    '''
    # Fitted pipeline
    fitted_pipeline = pipeline.fit(X, y)

    # Predictions
    y_pred = fitted_pipeline.predict(X)

    # Dataframe
    kncm = confusion_matrix(y, y_pred)
    kncm_df = pd.DataFrame(kncm, index=fitted_pipeline.classes_, columns=fitted_pipeline.classes_)
    kncm_df = kncm_df.unstack().reset_index()
    kncm_df.columns = ['Predicted', 'True', 'Count']
    kncm_df

    # Plot confusion matrix using plotnine
    plot = (
    ggplot(kncm_df, aes(x='Predicted', y='True', fill='Count')) +
    geom_tile(color='white') +
    geom_text(aes(label='Count'), size=10, color='black') +
    labs(title=f'Confusion Matrix - {model_name}', x='Predicted', y='True') + # the f is for f-string to include model name
    theme_minimal()
    )
    return plot
```

```python
# LDA score: 0.9258005281807546
lda_pipeline = Pipeline(
    [("lda", LinearDiscriminantAnalysis())]
)

ldascore = cross_val_score(lda_pipeline, X, y, cv=cv, scoring = 'roc_auc')
ldascore.mean()

ldafitted = lda_pipeline.fit(X,y)
ldapreds = ldafitted.predict(X)


# confusion matrix
confusion_matrix_pic('Linear Discriminant Analysis', lda_pipeline, X, y)
```

```python
# QDA score:0.7794697442815715
qda_pipeline = Pipeline([("qda",QuadraticDiscriminantAnalysis())])

qdascore = cross_val_score(qda_pipeline, X, y, cv=cv, scoring = 'roc_auc')
qdascore.mean()

ldafitted = lda_pipeline.fit(X,y)
ldapreds = ldafitted.predict(X)


# confusion matrix
confusion_matrix_pic('Quadratic Discriminant Analysis', qda_pipeline, X, y)
```

```python
# SVC score: 0.931296
svc_pipeline = Pipeline([('svc', SVC(kernel='linear', C = 0.01))])

c = [0.01,0.1,1,10,100]

c_values = {'svc__C':c}

gscv1 = GridSearchCV(svc_pipeline, c_values, cv = cv, scoring = 'roc_auc')
fitted_g = gscv1.fit(X,y)
results1 = pd.DataFrame(fitted_g.cv_results_)
results1


svc = pd.DataFrame(data = {"c": results1['param_svc__C'],"scores": results1['mean_test_score']})

svc.sort_values(by = 'scores', ascending = False)


# svcfitted = svc_pipeline.fit(X,y)
# svcpreds = svcfitted.predict(X)

# # confusion matrix
# confusion_matrix_pic('Support Vector Classifier', svc_pipeline, X, y)
```

```python
# SVM score: 0.933954
svm_pipeline = Pipeline([('svm', SVC(kernel='poly', C = 0.01, degree = 3))])


c = [0.01,0.1,1,10,100]
degree = [2, 4, 3, 5, 8, 10]
values = {'svm__C':c, 'svm__degree':degree}
gscv1 = GridSearchCV(svm_pipeline, values, cv = cv, scoring = 'roc_auc')
fitted_g = gscv1.fit(X,y)
results1 = pd.DataFrame(fitted_g.cv_results_)
results1


svm = pd.DataFrame(data = {"c": results1['param_svm__C'],"degree":results1['param_svm__degree'],"scores": results1['mean_test_score']})

svm.sort_values(by = 'scores', ascending = False)

svmfitted = svm_pipeline.fit(X,y)
svmpreds = svmfitted.predict(X)

# confusion matrix
confusion_matrix_pic('Support Vector Machines', svm_pipeline, X, y)
```

# Part Two: Natural Multiclass

My metrics were worse than in Part One becasue there is another category which is difficult to distinguish. The most categoires that were most likely to get mixed up were hybrid with any other type. Hybrid is a mix of both types which doesn't always provide clear classification compared to distinguishing between sativa and indica.

This is One v. Rest. So appropriate metrics to use in this multiclass setting is accuracy, precision, recall, and F1 Score.

Accuracy is the only metric that is truly a multiclass metric.

I did not consider micro/macro/weighted average because there is no specific interest in a singular category at this time.

```python
ct2 = ct.drop(['Strain','Effects','Flavor'], axis = 1)
X1 = ct2.drop(['Type'], axis = 1)

y1 = ct2['Type']
```

```python
# Decision Tree accuracy score: 0.622155
dtpipeline = Pipeline(
    [("dt", DecisionTreeClassifier(max_depth=3))
    ]
)

# Tuning

d = [1,3,5,7,10,20,30]

d_values = {'dt__max_depth':d}

gscvdt = GridSearchCV(dtpipeline, d_values, cv = cv, scoring = 'accuracy')
fitted_gdt = gscvdt.fit(X1,y1)
results1dt = pd.DataFrame(fitted_gdt.cv_results_)
results1dt

dt = pd.DataFrame(data = {"d": results1dt['param_dt__max_depth'],"scores": results1dt['mean_test_score']})
dt
dt.sort_values(by = 'scores', ascending = False)
```

```python
# Decision Tree plot

dtfitted = dtpipeline.fit(X1,y1)

# use this to get column names to use in feature names argument
feature_names = X1.columns.tolist()

# Export the decision tree to a DOT file
dot_data = export_graphviz(dtfitted.named_steps['dt'],
                           feature_names = feature_names,
                           class_names = dtfitted.named_steps['dt'].classes_.astype(str),
                           filled = True,
                           rounded = True,
                           special_characters = True,
                           out_file = None)

# Render the decision tree

graph = graphviz.Source(dot_data, format = 'png')
display(Image(graph.render(format='png', cleanup = True)))
```

Interpretation:

```python
# LDA accuracy score: 0.6234669678147939

hybridldapipeline = Pipeline(
    [('hlda',LinearDiscriminantAnalysis())]
)

hldascore = cross_val_score(hybridldapipeline, X1, y1, cv=cv, scoring = 'accuracy')
hldascore.mean()

# confusion matrix
confusion_matrix_pic('Linear Discriminant Analysis', hybridldapipeline, X1, y1)
```

```python
# QDA accuracy score: 0.20044795783926223

hybridqdapipeline = Pipeline(
    [('hqda', QuadraticDiscriminantAnalysis())]
)
hqdascore = cross_val_score(hybridqdapipeline, X1, y1, cv=cv, scoring = 'accuracy')
hqdascore.mean()
# confusion matrix
confusion_matrix_pic('Quadratic Discriminant Analysis', hybridqdapipeline, X1, y1)
```

```python
# kNN accuracy score: 0.610002

hybridknnpipeline = Pipeline(
    [('knn', KNeighborsClassifier(n_neighbors = 120))]
)

# Tuning

k = [1, 10, 30, 50, 70, 90, 100, 110, 120, 150]

k_values = {'knn__n_neighbors':k}

gscvknn = GridSearchCV(hybridknnpipeline, k_values, cv = cv, scoring = 'accuracy')
fitted_gknn = gscvknn.fit(X1,y1)
results1knn = pd.DataFrame(fitted_gknn.cv_results_)
results1knn

kNN = pd.DataFrame(data = {"k": results1knn['param_knn__n_neighbors'],"scores": results1knn['mean_test_score']})
kNN
kNN.sort_values(by = 'scores', ascending = False)

# confusion matrix
confusion_matrix_pic('KNN', hybridknnpipeline, X1, y1)
```
