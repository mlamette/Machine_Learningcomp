# Packages

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_selector, ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score, mean_squared_error
from plotnine import *
from sklearn.model_selection import GridSearchCV
```

# Data, Data Cleaning

```python
# Read the data
hitters = pd.read_csv("Hitters.csv")
hitters

# Drop other NAs
hitters = hitters.dropna()

hitters
```

# Part 1: Different Model Specs

## Regression without regularization

```python
# Pipeline - Ordinary Linear Regression

X = hitters.drop("Salary", axis = 1) # we drop Salary we want all columns as predictors
y = hitters["Salary"]


ct = ColumnTransformer(
  [("dummify",
    OneHotEncoder(sparse_output = False,handle_unknown='ignore'), # take it out and run it to see error
    make_column_selector(dtype_include=object)),
      ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))],
  remainder = "passthrough")


lr_pipeline_1 = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
).set_output(transform="pandas")
```

```python
# look at columns
dummy = ct.fit_transform(X)
dummy

# get columns and check length
col = list(dummy.columns)
len(col)
```

```python
# Fit
lr_fitted_1 = lr_pipeline_1.fit(X, y)
lr_fitted_1 # diagram

# Coefficients
coef_1 = lr_fitted_1.named_steps['linear_regression'].coef_

linear_coefficients = pd.DataFrame({'dependent_variable': col,'coefficients': coef_1})
linear_coefficients
```

Some of the coefficients that were negative that suggest that as they increase they have a negative effect on Salary. I find it interesting the as runs increases, Salary decreases. I would think that the more runs a person gets, the better effect on Salary.
AtBat, Runs, Walks, years
CRBI
League
Assists
New League

```python
# Cross val - the more negative the higher the error
scores = cross_val_score(lr_pipeline_1, X, y, cv = 5, scoring = 'neg_mean_squared_error')
-scores.mean()
```

### Mean Squared Error:
121136.31031816891

##Ridge regression

```python
# Pipeline 2 - Ridge Regression


ct2 = ColumnTransformer(
  [("dummify",
    OneHotEncoder(sparse_output = False,handle_unknown='ignore'), # take it out and run it to see error
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough")


lr_pipeline_2 = Pipeline(
  [("preprocessing", ct2),
  ("ridge_regression", Ridge(alpha = 1))]
).set_output(transform="pandas")
```

```python
# look at columns
dummy2 = ct2.fit_transform(X)
dummy2

# get columns and check length
col2 = list(dummy2.columns)
len(col2)
```

```python
# Fit
lr_fitted_2 = lr_pipeline_2.fit(X, y)
lr_fitted_2 # diagram

# Coefficients
coef_2 = lr_fitted_2.named_steps['ridge_regression'].coef_

ridge_coefficients = pd.DataFrame({'dependent_variable': col2,'coefficients': coef_2})
ridge_coefficients
```

League is no longer a negative coefficient

```python
# Tuning lambda aka alpha

# Your existing list of numbers
alphas = [0.001, 0.01, 0.1, 1, 10]


alpha_tune = {'ridge_regression__alpha': alphas} # need two underscores to separate name of step

gscv = GridSearchCV(lr_pipeline_2, alpha_tune, cv = 5, scoring='neg_mean_squared_error')

gscv_fitted = gscv.fit(X, y)

results = pd.DataFrame(gscv_fitted.cv_results_)
results

pd.DataFrame(data = {"alpha": results['param_ridge_regression__alpha'], "scores": -results['mean_test_score']})
```

The best MSE was for model with alpha is 1.

## Lasso Regression

```python
# Pipeline3 - Lasso
ct3 = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown='ignore'), # take it out and run it to see error
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lr_pipeline_3 = Pipeline(
  [("preprocessing", ct3),
  ("lasso", Lasso(alpha = 1))]
).set_output(transform="pandas")
```

```python
# look at columns
dummy3 = ct3.fit_transform(X)
dummy3

# get columns and check length
col3 = list(dummy3.columns)
len(col3)
# Fit on entire dataset
lr_fitted3 = lr_pipeline_3.fit(X, y)

# Estimating Betas for all variables - fit named steps
coef_3 = lr_fitted3.named_steps['lasso'].coef_

# Coefficients
pd.options.display.float_format = '{:.10f}'.format # removes scientific notation
lasso_coefficients = pd.DataFrame({'dependent_variable': col3,'coefficients': coef_3})
lasso_coefficients
```

```python
# Your existing list of numbers
alphas2 = [0.001, 0.01, 0.1, 1, 10]


alpha_tune2 = {'lasso__alpha': alphas} # need two underscores to separate name of step

gscv2 = GridSearchCV(lr_pipeline_3, alpha_tune2, cv = 5, scoring='neg_mean_squared_error')

gscv_fitted2 = gscv2.fit(X, y)

results2 = pd.DataFrame(gscv_fitted2.cv_results_)
results2

lasso1 = pd.DataFrame(data = {"alpha": results2['param_lasso__alpha'], "scores": -results2['mean_test_score']})
lasso1.sort_values(by = 'scores')
```

the alpha with the best MSE is alpha is 1

## Elastic Net

```python
# Pipeline 4 - Elastic Net

ct4 = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown='ignore'), # take it out and run it to see error
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lr_pipeline_4 = Pipeline(
  [("preprocessing", ct4),
  ("elastic", ElasticNet(alpha = 0.1, l1_ratio = 0.8))]
).set_output(transform="pandas")
```

```python

# look at columns
dummy4 = ct4.fit_transform(X)
dummy4

# get columns and check length
col4 = list(dummy4.columns)
len(col4)

# Fit on entire dataset
lr_fitted4 = lr_pipeline_4.fit(X, y)

# Estimating Betas for all variables - fit named steps
coef_4 = lr_fitted4.named_steps['elastic'].coef_


# Coefficients
pd.options.display.float_format = '{:.10f}'.format # removes scientific notation
elastic_coefficients = pd.DataFrame({'dependent_variable': col4,'coefficients': coef_4})
elastic_coefficients
```

```python
lambdas = [0.001, 0.01, 0.1, 1, 10]

alphas4 = [ 0.2, 0.4, 0.6, 0.8, 1]

alpha_tune4 = {'elastic__alpha': lambdas, 'elastic__l1_ratio':alphas4 } # need two underscores to separate name of step

gscv4 = GridSearchCV(lr_pipeline_4, alpha_tune4, cv = 5, scoring='neg_mean_squared_error')

gscv_fitted4 = gscv4.fit(X, y)

results4 = pd.DataFrame(gscv_fitted4.cv_results_)
results4


elastic= pd.DataFrame(data = {"alpha": results4['param_elastic__alpha'],"l1_ratio": results4['param_elastic__l1_ratio'] ,"scores": -results4['mean_test_score']})
elastic.sort_values(by = 'scores')
```

the model that had the lowest MSE was alpha is 0.01.

```python
# Part One Ordinary Coefficients
# Single Numeric
num_linear_coefficients = pd.DataFrame({'dependent_variable': col01,'coefficients': coef_01})
num_linear_coefficients['model']= 'linear_single'

# Five best
five_linear_coefficients = pd.DataFrame({'dependent_variable': col02,'coefficients': coef_02})
five_linear_coefficients['model'] = 'linear_five'

# Five best and interactions
five_inter_linear_coefficients = pd.DataFrame({'dependent_variable': col04,'coefficients': coef_04})
five_inter_linear_coefficients['model'] = 'linear_interactions'

# Part Two Ridge Coefficients
# Single Numeric
ridge_num_coefficients = pd.DataFrame({'dependent_variable': col11,'coefficients': coef_11})
ridge_num_coefficients['model'] = 'ridge_single'

# Five best
ridge_five_coefficients = pd.DataFrame({'dependent_variable': col12,'coefficients': coef_12})
ridge_five_coefficients['model'] = 'ridge_five'

# Five best and Interactions
ridge_five_interactions_coefficients = pd.DataFrame({'dependent_variable': col14,'coefficients': coef_14})
ridge_five_interactions_coefficients['model'] = 'ridge_interactions'

# stacks dataframes on top of each other
ridges = pd.concat([ridge_num_coefficients,ridge_five_coefficients,
           ridge_five_interactions_coefficients], ignore_index = True,
          axis = 0)
linears = pd.concat([num_linear_coefficients, five_linear_coefficients,
                     five_inter_linear_coefficients],
                    ignore_index = True,
                    axis = 0)
compare = pd.concat([ridges, linears], axis=1)
compare
```

- higher coefficient for ordinary linear than for ridge
- highest coefficient for five best variables for ridge was standardize hits
  - the highest for ordinary was also standardize hits
  - both also had AtBat as their second highest and negative
- highest coefficients for interaction variables for ridge was standardize hits and Division E
  - same for ordinary however it's coefficient was over 3x larger than ridge

All in all, the coefficients are relatively similary however you can see how the restrictions come in to play under ridge. The coefficients that are extremely large under the ordinary model are minimized under the ridge model.


## Lasso

```python
lasso1 = pd.DataFrame(data = {"alpha": results2['param_lasso__alpha'], "scores": -results2['mean_test_score']})
lasso_all = lasso1.sort_values(by = 'scores').head(1)
lasso_all['model'] = 'all_predictors'

scores21 = pd.DataFrame(data = {"alpha": results21['param_lasso_regression__alpha'], "scores": -results21['mean_test_score']})
lasso_single = scores21.sort_values(by = 'scores').head(1)
lasso_single['model'] = 'single_numeric'

scores22 = pd.DataFrame(data = {"alpha": results22['param_lasso_regression__alpha'], "scores": -results22['mean_test_score']})
lasso_five = scores22.sort_values(by = 'scores').head(1)
lasso_five['model'] = 'five_best'


scores24 = pd.DataFrame(data = {"alpha": results24['param_lasso_regression__alpha'], "scores": -results24['mean_test_score']})
lasso_interaction = scores24.sort_values(by = 'scores').head(1)
lasso_interaction['model'] = 'interactions'

lasso = pd.concat([lasso_all, lasso_single, lasso_five, lasso_interaction],
          axis = 0)
lasso
```

I got the same alpha result for the five best variables model with an alpha value of 1. The MSE is smallest for the models with smaller alpha values. This is because the larger the alpha, the more we care about restricting our coefficients to small values than about getting small MSE. This is why the MSE is larger for alpha 10. However, you can see that the interactions model has an MSE that isn't to far off from MSE of alpha 1 models yet it has an alpha value of 10.

Also within the interactions model, there were four interaction variables betas that equaled 0. This contributed to the Score being lower for the interaction model.

## Elastic Net Comparisons

```python
# Ridge
scores04 = pd.DataFrame(data = {"alpha": results['param_ridge_regression__alpha'], "scores": -results['mean_test_score']})
o_rid = scores04.sort_values(by='scores').head(1)
o_rid['model'] = 'all_predictors'

scores11 = pd.DataFrame(data = {"alpha": results11['param_ridge_regression__alpha'], "scores": -results11['mean_test_score']})
sing_rid = scores11.sort_values(by = 'scores').head(1)
sing_rid['model'] = 'single_numeric'

scores12 = pd.DataFrame(data = {"alpha": results12['param_ridge_regression__alpha'], "scores": -results12['mean_test_score']})
five_rid = scores12.sort_values(by = 'scores').head(1)
five_rid['model'] = 'five_best'

scores14 = pd.DataFrame(data = {"alpha": results14['param_ridge_regression__alpha'], "scores": -results14['mean_test_score']})
inter_rid = scores14.sort_values(by = 'scores').head(1)
inter_rid['model'] = 'interactions'

ridge = pd.concat([o_rid, sing_rid, five_rid, inter_rid], axis = 0).reset_index(drop = True)
ridge = ridge.rename(columns={"alpha": "r_alpha", "scores": "rid_scores", "model":"r_model"})

# Lasso
lasso = pd.concat([lasso_all, lasso_single, lasso_five, lasso_interaction],
          axis = 0).reset_index(drop = True)
lasso = lasso.rename(columns={"alpha": "l_alpha", "scores": "lass_scores", "model":"l_model"})

# Elastic
elastic= pd.DataFrame(data = {"alpha": results4['param_elastic__alpha'],"l1_ratio": results4['param_elastic__l1_ratio'] ,"scores": -results4['mean_test_score']})
o_elas = elastic.sort_values(by = 'scores').head(1)
o_elas['model'] = 'all_predictors'

scores31 = pd.DataFrame(data = {"alpha": results31['param_elastic__alpha'], "l1_ratio": results31['param_elastic__l1_ratio'],"scores": -results31['mean_test_score']})
sing_elas = scores31.sort_values(by = 'scores').head(1)
sing_elas['model'] = 'single_numeric'

scores32 = pd.DataFrame(data = {"alpha": results32['param_elastic__alpha'],"l1_ratio": results32['param_elastic__l1_ratio'], "scores": -results32['mean_test_score']})
five_elas = scores32.sort_values(by = 'scores').head(1)
five_elas['model'] = 'five_best'

scores34 = pd.DataFrame(data = {"alpha": results34['param_elastic__alpha'],"l1_ratio": results34['param_elastic__l1_ratio'] ,"scores": -results34['mean_test_score']})
inter_elas = scores34.sort_values(by = 'scores').head(1)
inter_elas['model'] = 'interactions'

elastic = pd.concat([o_elas, sing_elas, five_elas, inter_elas],
                    axis = 0).reset_index(drop = True)
elastic = elastic.rename(columns={"alpha": "e_alpha", "scores": "elas_scores", "model":"e_model"})

# put tables together for comparison
compare2 = pd.concat([ridge, lasso, elastic], axis=1)
compare2
```

Elastic had the smalles MSE across the board. It makes sense that Elastic wins every time because it is ridge penalty and lasso penalty and combines using both alpha's and lambdas. So we have a weight assigned to Ridge vs. Lasso and weight assigned to SSE and penalties.
